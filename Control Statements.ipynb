```python
#WAP to check whether the number is Armstrong or not.

n = int(input("Enter the number: "))
m=n
sum=0
while n>0:
    a=n%10
    sum+=a**len(str(m))
    n=n//10
    print(a)
print('Sum of cube of each number present in the number: ',sum)
if sum==m:
    print(f"The given number '{m}' is a Armstrong Number")
else:
    print("Not a Armstrong number.")
```

```python
#WAP to find the HCF of two numbers

n=int(input("Enter the first number: "))
m=int(input("Enter the second number: "))
hcf=1
i=1
while i<=n and i<=m:
    if n%i==0 and m%i==0:
        hcf=i
    i+=1
print(hcf)
```

```python
#WAP to check the given number is prime or not.
n=int(input("Enter the number: "))
i=2
output=1
if n>1:
    while i<n:
        if n%i==0:
            output=0
        i+=1
    if output==1:
        print("Prime Number")
    else:
        print("Not a prime number")
else:
    print("Not a prime number.")
```

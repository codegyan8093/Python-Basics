```python
s1={10,20,30,40}
s2={30,40,50,60}
print(s1.union(s2))# aiso use | in place of union
print(s1.intersection(s2))# aiso use & in place of intersection
print(s1.difference(s2))# aiso use - in place of difference
print(s1.symmetric_difference(s2))# aiso use ^ in place of symmetric_difference
```

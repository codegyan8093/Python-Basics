# Dictionary
  d={111:123,222:231,333:321}
  duplicates keys are not allowed but values can be duplicated
  if we try to add duplicate keys then old value changed by new value
  order is not preserved all element inserted accoding to hash of key
  indexing and slicing not allowed
# how to create dict
  d={}
  d=dict()
  d[key]=value-->the provided key and value added to the dict   
# Accessing data in dict
  by using the key we access the corresponding value  
# important method() and fun()
  dict()-->to create dict
  get(key)-->return the corresponding value, if not present we will get None   
  pop(key)-->remove the item associate with specific key
  popitem()-->remove rendom key value pair
  keys()-->gives only keys
  values()-->gives only values
  items()-->gives key value pair
  copy()-->to duplicate dictionary
  update()-->all the items in x will be added in dictionary d
